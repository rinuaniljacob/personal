```python
#!pip install crewai
from crewai import Agent, Task, Crew, Process
# Initialize tools
from crewai_tools import ScrapeWebsiteTool, SerperDevTool

# Initialize the tools
search_tool = SerperDevTool()
scrape_tool = ScrapeWebsiteTool()
```

```python
# Industry Researcher Agent
industry_researcher = Agent(
    role="Senior Industry Researcher",
    goal="Conduct and indepth industry and market research on {company_name}.",
    backstory=(
        "You are a senior market researcher with 25 years of experience analyzing industries and identifying key focus areas for companies. "
"Your expertise spans sectors like Automotive, Manufacturing, Finance, Retail, and Healthcare. "
"You specialize in understanding company vision, operations, supply chains, and customer experience, "
"and your research lays the foundation for AI/ML solutions tailored to {company_name}."       
    ),
    #llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python

# Use Case Analyst Agent
use_case_analyst = Agent(
    role="Senior AI/ML solutions architect",
    goal="Propose relevant AI use cases where {company_name} can leverage GenAI, LLMs, and ML technologies.",
    backstory=(
        "You are a creative AI solutions expert with 20 years of experience identifying and implementing AI/ML technologies. "
"Building on industry research, you craft innovative use cases that solve company pain points and enhance efficiency. "
"You specialize in leveraging GenAI, ML, and automation to align business processes with long-term goals."

    ),
    #llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
# Resource Collector Agent
resource_collector = Agent(
    role="Resource Asset Collection Researcher",
    goal="Collect datasets and resources for implementing the AI use cases proposed for {company_name}.",
    backstory=(
    "You are a skilled researcher and resource curator with expertise in finding datasets and tools on platforms like Kaggle, HuggingFace, and GitHub. "
"Your work ensures actionable resources for implementing AI/ML use cases proposed by the solutions architect."
),
    #llm=llm,
    verbose=True,
    allow_delegation=False
)
```

```python
# Proposal Creator Agent
proposal_creator = Agent(
    role="Proposal Creator",
    goal="Generate a detailed proposal with insights, use cases, and resource links for {company_name}.",
    backstory=(
"You are an articulate professional who creates actionable proposals by synthesizing insights from researchers, solutions architects, and resource curators. "
"Your focus is on delivering high-quality, client-ready recommendations with clarity and precision."

    ),
    #llm=llm,
    allow_delegation=True,
    verbose=True,    
)
```

```python

# Tasks
research_task = Task(
    description=(
        "Analyze {company_name} and its industry. Identify strategic focus areas, market trends, competitors, and key players. "
"Provide an overview of the company's vision, product offerings, and segment priorities."
    ),
    expected_output="A markdown report with company and industry insights and research.",
    agent=industry_researcher,
    tools=[search_tool,scrape_tool],
    output_file="industry_research.md",
)
```

```python
use_case_task = Task(
    description=(
        "Analyze research findings to propose at least 10 AI/ML/GenAI use cases for {company_name}."
        "Refer to reports and insights on AI and digital transformation from industry-specific sources such as McKinsey, Deloitte, or Nexocode."
        "Identify pain points and threats, "
"refer to industry reports for insights, and ensure feasibility and practicality. Proposals should include applications to enhance processes, "
"boost operational efficiency, and improve customer satisfaction."

    ),
    expected_output="A markdown file listing AI use cases with explanations, include atleast 10 use cases.",
    agent=use_case_analyst,
    tools=[search_tool,scrape_tool],
    output_file="ai_usecases.md",
)
```

```python
resource_task = Task(
    description=(
        "Search platforms like Kaggle, HuggingFace, and GitHub to find relevant datasets for the proposed use cases. "
"Ensure all resources are specific to {company_name}'s industry and provide clickable links."

    ),
    expected_output="A markdown file containing resource links for datasets and tools.",
    input_file="ai_usecases.md",
    agent=resource_collector,
    tools=[search_tool,scrape_tool],
    output_file="resources.md",
)
```

```python

proposal_task = Task(
    description=(
        "Compile a final proposal with research insights, at least 10 AI use cases, and resource links. "
"Begin with an overview of how {company_name} can leverage Generative AI, LLMs, and ML to improve efficiency, product quality, and service offerings. "
"Each use case must include the problem, proposed AI solution, benefits, and clickable references/datasets where available. "
"Ensure the proposal is actionable, professional, and client-focused and make sure to list atleast 10 use cases in the proposal."
"The crucial aspects are the depth of market research and competitor analysis,"
"relevance and creativity of use cases generated for the customer, feasibility and practicality of implementing the proposed solutions, "
"Quality and clarity of the final proposal, including actionable insights."
 
    ),
    expected_output="A markdown file with the final detailed proposal for the company with list of atleast 10 use cases.",
    agent=proposal_creator,
    output_file="final_proposal.md",
)
```

```python
crew = Crew(
    agents=[industry_researcher, use_case_analyst, resource_collector, proposal_creator],
    tasks=[research_task, use_case_task, resource_task, proposal_task],
    process=Process.sequential,
    memory=True,
    verbose=2
)
```

```python
# Kickoff the Crew
inputs = {"company_name": "SpaceX"}
result = crew.kickoff(inputs=inputs)
print(result)
```
